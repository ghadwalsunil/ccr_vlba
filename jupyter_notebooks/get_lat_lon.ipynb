Using Google Map Static API

```python
import requests
import config
api_key = config.API_KEY

url = f"https://maps.google.com/maps/api/staticmap?center=Garbsen,&zoom=15&markers=color:red%7Clabel:C%7C52.4276506,9.6005202&size=640x640&key={api_key}&maptype=roadmap"
r = requests.get(url)
print(r.content)
```
