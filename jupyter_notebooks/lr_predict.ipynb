```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('lr.csv')
df
```

```python
df['Total'] = df['Cost'] + df['Salary'] + df['TotalHotelCost'] + df['TotalTravelCosts'] + df['TotalEquipmentCost']
df.head()
```

```python
train_df = df[(df['Year'] >= 2000) & (df['Year']<2021)]
train_df
```

```python
X_train = df.iloc[:,2:-1].values
y_train = df.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
avg_df = df[df['Year'] == 2021]
avg_df
```

```python
y = avg_df.mean()
y
```

```python
X_test = [[y.Cost,y.Salary,y.TotalHotelCost,y.TotalTravelCosts,y.TotalEquipmentCost]]
y_pred = lr.predict(X_test)
```
