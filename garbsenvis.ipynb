```python
def get_lat_lon(col):
    location = geolocator.geocode(col)
    new = location.raw["lat"]+ "," + location.raw["lon"]
    return new
```

```python
cost_zero_df = df[df['UHC Garbsen'] == 0]
cost_zero_df['lat_lon'] = cost_zero_df['City_x'].apply(get_lat_lon)
cost_zero_df.head()
```
