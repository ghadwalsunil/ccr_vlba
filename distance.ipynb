```python
import requests
import pandas as pd
import json
```

```python
df_trips = pd.read_csv("./data_CCR/CCR_2/MwwCcrBusinessTrips_2.csv", encoding = 'unicode_escape',on_bad_lines='skip')
df_branch = pd.read_csv("./data_CCR/CCR_2/MwwCcrBranches_2.csv", encoding = 'unicode_escape',on_bad_lines='skip')
df_emp = pd.read_csv("./data_CCR/CCR_2/MwwCcrEmployees_2.csv", encoding = 'unicode_escape',on_bad_lines='skip')
```

```python
df_join = df_trips.merge(right=df_emp, how="inner", on="EmployeeId").merge(right=df_branch, how="inner", on="BranchId")
df_join = df_join[(df_join["HotelCosts"] > 0) & ((df_join["Position"] == "Consultant"))]
df_join
```

```python
df_join = df_join[["PostalCode_x", "City_x"]]#, "PostalCode_y", "City_y"]]
df_join.columns = ["Dest_PC", "Dest_City"]#, "Branch_PC", "Branch_City"]
df_join
```

```python
df_join = df_join.drop_duplicates().copy()
df_join
```

```python
# df_join["Branch"] = df_join["Branch_PC"].astype("str")
df_join["Dest"] = df_join["Dest_PC"].astype("str")
```

```python
# df_join["Branch"] = df_join["Branch"] + "%2C%20" + df_join["Branch_City"]
df_join["Dest"] = df_join["Dest"] + "%2C%20" + df_join["Dest_City"]
```

```python
df_join = df_join.drop(df_join[(df_join['Dest_City'] == 'Garbsen') | (df_join['Dest_City'] == 'Hannover')].index)
df_join
```

```python
df_join["ResponseModGarbsen"] = df_join["Response Garbsen"].apply(lambda x: json.loads(x))
df_join["ResponseModHannover"] = df_join["Response Hannover"].apply(lambda x: json.loads(x))
df_join
```

```python
def get_value(value, param):
    try:
        return value["rows"][0]["elements"][0][param]["text"]
    except:
        return "NOT_AVAILABLE"
```

```python
df_join["DistanceGarbsen"] = df_join["ResponseModGarbsen"].apply(lambda x: get_value(x, "distance"))
df_join["DurationGarbsen"] = df_join["ResponseModGarbsen"].apply(lambda x: get_value(x, "duration"))
df_join
```

```python
df_join["DistanceHannover"] = df_join["ResponseModHannover"].apply(lambda x: get_value(x, "distance"))
df_join["DurationHannover"] = df_join["ResponseModHannover"].apply(lambda x: get_value(x, "duration"))
df_join
```

```python
df_join = df_join.drop(df_join[(df_join['DistanceHannover'] == 'NOT_AVAILABLE') | (df_join['DurationHannover'] == 'NOT_AVAILABLE') | (df_join['DistanceGarbsen'] == 'NOT_AVAILABLE') | (df_join['DurationGarbsen'] == 'NOT_AVAILABLE')].index)
df_join
```

```python
def calc_minutes(col):
    x = col.split(' ')
    if len(x) == 4:
        minutes = (int(x[0]) *60 ) + int(x[2])
    else:
        minutes = int(x[0])
    return minutes
```

```python
df_join['Garbsen_minutes'] = df_join['DurationGarbsen'].apply(calc_minutes)
print(df_join)
```

```python
df_join['Hannover_minutes'] = df_join['DurationHannover'].apply(calc_minutes)
print(df_join)
```
